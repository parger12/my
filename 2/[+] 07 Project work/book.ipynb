### Шаг 2.2 Проверка данных на аномалии и исправления.

```python
data ['days_employed'] = data['days_employed'].abs()
days_employedd_median = data['days_employed'].median()
data ['days_employed'] = data['days_employed'].fillna(days_employedd_median)
print (data.info())
```

### Шаг 2.3. Изменение типов данных.

```python
data['total_income'] = data['total_income'].astype(int)
print(data.info())
```

### Шаг 2.4. Удаление дубликатов.

```python
data = data.drop_duplicates().reset_index(drop=True)

data['education'] = data['education'].str.lower()

print (data.head(10))
```

### Шаг 2.5. Формирование дополнительных датафреймов словарей, декомпозиция исходного датафрейма.

```python
education_dict = data[['education_id', 'education']].drop_duplicates().reset_index(drop=True)
family_status_dict = data[['family_status', 'family_status_id']].drop_duplicates().reset_index(drop=True)

data = data.drop(columns=['education', 'family_status'])

print (data)
```

### Шаг 2.6. Категоризация дохода.

```python
def income_total(row):
    if 0 < row < 30000:
        return 'E'
    elif 30001 < row < 50000:
        return 'D'
    elif 50001 < row < 200000:
        return 'C'
    elif 200001 < row < 1000000:
        return 'B'
    elif row > 1000001:
        return 'A'
    else:
        return 'Неизвестная категория'
    

data['total_income_category'] = data['total_income'].apply(income_total)

print (data.head(10))
```

### Шаг 2.7. Категоризация целей кредита.

```python
def purpose_kat (row):
    if 'жиль' in row or 'недвижи' in row:
        return 'операции с недвижимостью'
    elif 'авто' in row:
        return 'операции с автомобилем'
    elif 'образов' in row:
        return 'получение образования'
    elif 'свадьб' in row:
        return 'проведение свадьбы'
    
data['purpose_category'] = data['purpose'].apply(purpose_kat)

print(data['purpose_category'].value_counts())
    

    
```
